```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
#library imports
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
import re
import spacy
from collections import Counter
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F
import string
from torch.nn.utils.rnn import pack_padded_sequence, pad_packed_sequence
from sklearn.metrics import mean_squared_error
import re
from sklearn.preprocessing import MinMaxScaler
```

```python
import pandas as pd
data = pd.read_csv('/content/task_1_information_extraction_train.tsv', sep = '\t')
print(len(data))
```

```python
data=data[:10000]
```

```python
# Removies Null rows and reseting Index
data = data[data['NOTES'].notna()]
data = data.reset_index(drop=True)
# data = data[:20000]
```

```python
newdata = pd.concat([data['NOTES'], data['INTER2']], axis=1)
```

```python
tok = spacy.load('en')
def tokenize (text):
    text = re.sub(r"[^\x00-\x7F]+", " ", text)
    regex = re.compile('[' + re.escape(string.punctuation) + '0-9\\r\\t\\n]') # remove punctuation and numbers
    nopunct = regex.sub(" ", text.lower())
    return [token.text for token in tok.tokenizer(nopunct)]
```

```python
counts = Counter()
docs = {}
for i in range(len(newdata['NOTES'])):
    docs[i] = tokenize(newdata['NOTES'][i])
    counts.update(tokenize(newdata['NOTES'][i]))
```

```python
#creating vocabulary
vocab2index = {"":0, "UNK":1}
words = ["", "UNK"]
for word in counts:
    vocab2index[word] = len(words)
    words.append(word)
```

```python
def encode_sentence(text, vocab2index, N=70):
    tokenized = tokenize(text)
    encoded = np.zeros(N, dtype=int)
    enc1 = np.array([vocab2index.get(word, vocab2index["UNK"]) for word in tokenized])
    length = min(N, len(enc1))
    encoded[:length] = enc1[:length]
    return encoded, length
```

```python
newdata['encoded'] = newdata['NOTES'].apply(lambda x: np.array(encode_sentence(x,vocab2index)))
```

```python
X_train = []
for i in range(len(newdata['encoded'])):
    X_train.append(newdata['encoded'][i][0])
```

```python
target = []
for i in range(len(newdata['INTER2'])):
    target.append(newdata['INTER2'][i])
```

```python
X_train = np.array(X_train)
print(X_train.shape)
target = np.array(target)
```

```python

# sc = MinMaxScaler(feature_range = (0, 1))
# training_set_scaled = sc.fit_transform(X_train)

X_train1 = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
# Inter1
regressor = Sequential()

regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))

regressor.add(Dense(units = 1))

regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

regressor.fit(X_train1, target, epochs = 500, batch_size = 32)
```
